```python
from horseshoe_gibbs import sample_gibbs_improved
import torch
import matplotlib.pyplot as plt
```

```python
n = 5
p = 10
max_mcs = 1000

torch.manual_seed(1)
w = torch.randn(p)

X = torch.bernoulli(0.5 * torch.ones(n, p))
X = torch.vstack([X for _ in range(100)])
y = X @ w

print(X.shape, y.shape)
```

```python
betas, sigma2s, tau2s, lamb2s = sample_gibbs_improved(
    X, y, max_mcs, n_warmup=0, debug=True
)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(6, 6), dpi=200)
fig.subplots_adjust(hspace=0.1, wspace=0.1)

axes[0].plot(betas)
axes[1].plot(sigma2s)
axes[2].plot(tau2s)
```
